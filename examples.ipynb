## Random Variables

**Definition.** Let e be an experiment and $\mathcal{S}$ a sample space associated with the experiment. A function $X$ assigning to every element $s \in \mathcal{S}$, a real number, $X(s)$, is called a random variable.

**Example 1:** Consider the following random variable generated by a dice $\mathcal{S}=\{1,2,3,4,5,6\}$:
- $X=+\$3$ if $s=\{1,2\}$
- $X=-\$2$ if $s=\{3,4\}$
- $X=+\$1$ if $s=\{5\}$
- $X=-\$4$ if $s=\{6\}$
 
1. Compute the expectation (mean) of $X$.
2. Compute the variance of $X$. 
3. Plot the distribution of this random variable.

```python

## Example 1
import matplotlib.pyplot as plt
X_values = ["+$3","-$2","+$1","-$4"]
proba=[2/6,2/6,1/6,1/6]
# Plotting the probability distribution
plt.figure(figsize=(8, 5))
bars = plt.bar(X_values, proba, color=['green', 'red', 'blue', 'purple'])

# Adding labels and title
plt.xlabel('Values of X')
plt.ylabel('Probability')
plt.title('Probability Distribution of the Random Variable X')

# Display the plot
plt.show()
```

```python
import numpy as np
```

```python
expected_value = proba[0]*(3)+proba[1]*(-2)+proba[2]*(1)+proba[3]*(-4)
print(expected_value)
```

**Example 2:** Suppose that items coming off a production line are classified as defective (D) or nondefective (N). Suppose that three items are chosen at
random from a day's production and are classified according to this scheme. Let us suppose that with probability 0.1 an item is defective and hence with
probability 0.9 an item is nondefective.
 
1. Find the Sample Space for this experiment.
2. Compute the probability distribution
3. Plot the distribution of this random variable.

We have a profit of $2 for each nondefective item and a loss of -$3 for each defective item:

 - Compute the probability distribution of this new random variable.
 - Compute the expected profit/loss.
 - How much should we decrease the defective to have a positive profit ?


```python
## Example 2
import itertools
# Elements
elements = ['D', 'N']

# Using itertools.product to get all combinations of 3 elements
sample_space = list(itertools.product(elements, repeat=3))
print(sample_space)
```

```python
# Values of the random variable X (number of defective items)
X_values = [0, 1, 2, 3]

# Probabilities of each value of X
probabilities = [
    1*(0.9)**3,             # P(X=0)
    3 * (0.9)**2 * 0.1,  # P(X=1)
    3 * 0.9 * (0.1)**2,  # P(X=2)
    1*(0.1)**3             # P(X=3)
]
print(probabilities)
```

```python
expected_value=probabilities[0]*(3*2)+probabilities[1]*(2*2-1*3)+probabilities[2]*(1*2-2*3)+probabilities[3]*(0*2-3*3)
print(expected_value)
```

```python
# Plotting the probability distribution
plt.figure(figsize=(8, 5))
plt.bar(X_values, probabilities, color=['blue', 'green', 'red', 'purple'])

# Adding labels and title
plt.xlabel('Number of Defective Items in a Selection of 3')
plt.ylabel('Probability')
plt.title('Probability Distribution of the Number of Defective Items')

# Display the plot
plt.show()
```
